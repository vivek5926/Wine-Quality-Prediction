# Feature Selection

```python
# Create Classification version of target variable
wine['goodquality'] = [1 if x >= 7 else 0 for x in wine['quality']]# Separate feature variables and target variable
X = wine.drop(['quality','goodquality'], axis = 1)
Y = wine['goodquality']
```

```python
# See proportion of good vs bad wines
wine['goodquality'].value_counts()
```

```python
X
```

```python
print(Y)
```

# Feature Importance

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

from sklearn.ensemble import ExtraTreesClassifier
classifiern = ExtraTreesClassifier()
classifiern.fit(X,Y)
score = classifiern.feature_importances_
print(score)
```

# Splitting Dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state=7)
```

# LogisticRegression:

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,Y_train)
Y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print("Accuracy Score:",accuracy_score(Y_test,Y_pred))
```

```python
confusion_mat = confusion_matrix(Y_test,Y_pred)
print(confusion_mat)
```

```python
TP = confusion_mat[1,1] # true positive 
TN = confusion_mat[0,0] # true negatives
FP = confusion_mat[0,1] # false positives
FN = confusion_mat[1,0] # false negatives
```

```python
# Sensitivity
print("Sensitivity:-",TP / float(TP+FN))

# Specificity
print("Specificity:-", TN / float(TN+FP))
```

```python
# classification_report
from sklearn.metrics import classification_report
print(classification_report(Y_test, Y_pred))
```

```python
# Predict the quality of wine using our model (predict())
Y_train = model.predict(X_train)
print(Y_train)
Y_pred = model.predict(X_test)
print(Y_pred)
```
